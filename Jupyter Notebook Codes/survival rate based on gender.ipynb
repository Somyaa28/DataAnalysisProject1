```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df=sns.load_dataset('titanic')
df
```

```python
df.isnull().sum()
```

```python
df=df.dropna(subset=['embarked','embark_town'])
df
```

```python
df[df.duplicated()]
```

```python
df.isnull().sum()
```

```python
df=df.drop_duplicates()
df
```

```python
age=df["age"].groupby([df["who"], df["sex"]]).agg('mean')
age
```

```python
df['age']=df["age"].groupby([df["who"], df["sex"]]).transform(lambda x : x.fillna(x.mean()))
df
```

```python
sns.countplot(x='who',hue='alive',data=df)
```
